# Neue Zurcher Zeitung

(mind the gap after 1950 - they apparently doesn't show the results for those!

```python
import json
import pandas as pd

rows = []
with open("nzz-counts.impresso", "r") as f:
    for line in f.readlines():
        rows.append(json.loads(line))
```

```python
nzz_df = pd.DataFrame(rows[0][1][0])
nzz_df = nzz_df.buckets.apply(lambda row: pd.Series({'year': row['val'], 'article_count': row['count']}))
nzz_df = nzz_df.sort_values(by='year').set_index('year')
nzz_df.plot()
```
